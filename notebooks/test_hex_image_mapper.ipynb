## Imports

```python
#Importing ImageMapper
from ctlearn.image_mapping import ImageMapper


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating the Mappers

```python
'''This iPython notebook will print out the events with the help of a toy dataset for the two hex conversion algos 
image shifting and axial addressing. It will do so for all the hex_cams with various amounts of padding. 
This process will then be repeated for a few test images from the MAGICcam.
'''

#Load various (hex) Image Mappers with various arguments for testing purposes

#Cameras with hex output
hex_cams = ['LSTCam', 'FlashCam', 'NectarCam', 'DigiCam', 'VERITAS','MAGICCam', 'FACT', 
            'HESS-I', 'HESS-II'] 

#Testing the 'image shifting' and 'axial addressing' conversion algorithms explicitly for the hex_cams
hex_algos = ['image_shifting', 'axial_addressing']


#Creating a dictionary with key as hex_algo and value as a list of different Image Mappers with that algo

mappers = {hex_algo: [] for hex_algo in hex_algos} 
                                                    
for hex_algo in hex_algos:
    print(hex_algo)
    hex_conversion_algorithm = {cam: hex_algo for cam in hex_cams}
    print(hex_conversion_algorithm)
    #various amounts of padding
    mappers[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm))
    mappers[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm,padding={cam: 5 for cam in hex_cams}))
    mappers[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm,padding={cam: 10 for cam in hex_cams}))
    mappers[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm,padding={cam: 15 for cam in hex_cams}))
```

```python
print(mappers['image_shifting'])
print(mappers['axial_addressing'])
```

## Creating the toy datasets

```python
#creating a 2 toy datasets, with different pixel values 

test_pixel_values = {}
test_pixel_values2 = {}
for tel in hex_cams:
    test_pixel_values[tel] = np.concatenate(([0.0], np.arange(0, mappers['image_shifting'][0].num_pixels[tel], 1)))
    test_pixel_values2[tel] = np.concatenate(([0.0], (np.arange(mappers['image_shifting'][0].num_pixels[tel],0,-1))))
    
    
    test_pixel_values[tel] = np.expand_dims(test_pixel_values[tel], axis=1)
    print(test_pixel_values[tel].shape, tel)
    test_pixel_values2[tel] = np.expand_dims(test_pixel_values2[tel], axis=1)
    #print("second",test_pixel_values2[tel])
```

```python
#Taken from test_image_mapper, same color scheme

def plot_image(mapper, camera_type):
    print(camera_type)
    image = mapper.map_image(test_pixel_values[camera_type], camera_type)
    fig, ax = plt.subplots(1)
    ax.set_aspect(1)
    ax.pcolor(image[:,:,0], cmap='viridis')
    plt.show()
    
def plot_image2(mapper, camera_type):
    print(camera_type)
    image = mapper.map_image(test_pixel_values2[camera_type], camera_type)
    fig, ax = plt.subplots(1)
    ax.set_aspect(1)
    ax.pcolor(image[:,:,0], cmap='viridis')
    plt.show()
```

## Printing resultant images

```python
for cam in hex_cams:
    for hex_algo in hex_algos:
        for i in range(len(mappers[hex_algo])):
            print("Padding = ", i*5)
            plot_image(mappers[hex_algo][i], cam)
            plot_image2(mappers[hex_algo][i],cam)    
```

```python
def plot_magic_image(mapper, camera_type, test_data_magic):
    #print(camera_type)
    #print(test_data_magic)
    image = mapper.map_image(test_data_magic[camera_type], camera_type)
    fig, ax = plt.subplots(1)
    ax.set_aspect(1)
    ax.pcolor(image[:,:,0], cmap='viridis')
    plt.show()
```

## Creating the mappers

```python
test_data_magic = {}
magicmapper = {hex_algo: [] for hex_algo in hex_algos}

#using different parameters like padding, mask_interpolation and peak_times
for hex_algo in hex_algos:
    print(hex_algo)
    hex_conversion_algorithm = {'MAGICCam': hex_algo}
    print(hex_conversion_algorithm)
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm, use_peak_times=True))
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm, use_peak_times=False))
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm,padding={cam: 5 for cam in hex_cams}))
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm,padding={cam: 10 for cam in hex_cams}))
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm, mask_interpolation=True))
    magicmapper[hex_algo].append(ImageMapper(hex_conversion_algorithm=hex_conversion_algorithm, mask_interpolation=False))
    
```

```python
labeldictionary = {0: 'use_peak_times = True', 1: 'use_peak_times = False', 2: 'Padding = 5', 3: 'Padding = 10', 4: 'mask_interpolation = True', 5: 'mask_interpolation = False'}
```

## Printing out results

```python
for idx,row in enumerate(data.root.MAGIC.iterrows()):
    if (idx==0):     #Since the first test image is blank and consists of only zeros
        continue
    test_data_magic['MAGICCam'] = np.concatenate(([0.0], np.array(row['image_charge'])))
                                                 
    test_data_magic['MAGICCam'] = np.expand_dims(test_data_magic['MAGICCam'], axis=1)
    print(test_data_magic['MAGICCam'].shape)
    for hex_algo in hex_algos:
        for i in range(len(magicmapper[hex_algo])):
            print("Algo: ", hex_algo)
            print(labeldictionary[i])
            plot_magic_image(magicmapper[hex_algo][i], 'MAGICCam',test_data_magic)
```
